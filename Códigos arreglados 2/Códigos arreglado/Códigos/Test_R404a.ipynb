# Referencia para ciclo en cascada


Antes de llamar al ciclo y resolver, se deben fijar dos cosas: Condiciones fijas a las cuales el sistema operar, y variables que se definen en el sistema. Con ello, se puede empezar la optimizacin del sistemas para cada refrigerante.

### Supuestos:

Se resuelve el ciclo de forma termodinmica; los parmetros como potencia entregada por el compresor, prdida de carga en dispositivo de expansin, entre otros, se pueden ajustar en base al ciclo. Para ello, s se deben tener en cuenta equipos reales que alcancen valores cercanos a los requeridos, o tengan potencia/revoluciones variables (inverter).

```python
HX_ev_low = HX(T_c_in=T_e_low, m_h=m_g_e_low, U=U_e_low, pinch=pinch_ev_low, Np=Np, Nt=Nt, di=di, t=t, fluid_c=fluid_low, fluid_h=glicol, tipo='evaporador')
HX_cond_high = HX(T_c_in=T_amb, U=U_c_high, pinch=pinch_ev_low, Np=Np, Nt=Nt, di=di, t=t, fluid_h=fluid_high, fluid_c=aire, tipo='condensador')


ciclo_combinado = VCR_cascada(Q=Q, T_ev_low=T_e_low, T_cond_low=T_c_low, T_ev_high=T_e_high, T_cond_high=T_c_high,
                              ref1=fluid_low, ref2=fluid_high, eta_c_low=eta_s, eta_c_high=eta_s,
                              Tsh=Tsh, Tsc=Tsc, T_amb=T_amb, ref_e_low=glicol, U_e_low=U_e_low, T_chamber=T_chamber,
                              m_g_e_low=m_g_e_low, m_aire=m_aire, HX_e_low=HX_ev_low, HX_c_high=HX_cond_high)

m_c = ciclo_combinado.m_dot_low

ciclo_combinado.ciclo()
ciclo_combinado.show()
```

```python
from pymoo.core.problem import ElementwiseProblem
from pymoo.algorithms.moo.nsga2 import NSGA2
from pymoo.optimize import minimize
from pymoo.visualization.scatter import Scatter
from pymoo.algorithms.soo.nonconvex.cmaes import CMAES
from pymoo.algorithms.soo.nonconvex.ga import GA
from pymoo.core.mixed import MixedVariableGA
from pymoo.algorithms.moo.nsga2 import RankAndCrowdingSurvival
from pymoo.termination import get_termination

# -------------------------------------------------
# Funcin que monta el sistema y devuelve mtricas
# -------------------------------------------------
def componentes(pinch_ev, pinch_cond, T_dif):
    #Estos valores son fijos
    T_chamber = -25 + 273.15 #K
    T_out = 25 + 273.15 #K
    Q = 200e3 #W a enfriar


    eta_s = 0.8 #Eficiencia isentrpica
    Tsc = 5 #K, subrenfriamiento en condensador
    Tsh = 5 #k, sobrecalentamiento en evaporador
    T_amb = 25 + 273.15 #K, temperatura ambiente


    T_e_low = T_chamber - pinch_ev
    T_c_low = T_e_low + T_dif #Se fija T del condensador del primer ciclo para que la temperatura sea 30 grados ms que en el evaporador

    T_e_high = T_c_low - 10 #De tal manera que hayan 10 K de diferencia en el intercambiador
    T_c_high = T_out + pinch_cond #Temperatura es mayor que el ambiente


    fluid_low = 'R41'
    fluid_high = 'R41'

    #Y para el evaporador de baja temperatura
    m_g_e_low = 1.5 #kg/s
    rho_e_low = 850 #kg/m3
    cp_e_low = 2100 #J/kgK

    k_h = 0.13 #W/mk
    mu_h = 3.5*10**(-3) #Pa*s

    glicol = refrigerante(cp=cp_e_low, rho=rho_e_low, k=k_h, mu=mu_h) #Fluido caliente en el ev. para el ciclo de baja T
    aire = refrigerante(cp=1000, rho=1, k=0.024, mu=1.81*10**(-5)) #Fluido fro en el cond. del ciclo de alta T

    m_aire = 5 #kg/s

    U_e_low = 800 #Caso base
    U_c_high = 800 #Caso base

    Np = 2 #Nmero de pasos
    Nt = 80 #Nmero de tubos
    di = 0.01 #m, diametro interno
    t = 0.002 #m, espesor

    try:
        # crear intercambiadores usando los pinch dados
        HX_ev_low = HX(T_c_in=T_e_low, m_h=m_g_e_low, U=U_e_low, pinch=pinch_ev,
                       Np=Np, Nt=Nt, di=di, t=t,
                       fluid_c=fluid_low, fluid_h=glicol, tipo='evaporador')

        HX_cond_high = HX(T_c_in=T_amb, U=U_c_high, pinch=pinch_cond,
                          Np=Np, Nt=Nt, di=di, t=t,
                          fluid_h=fluid_high, fluid_c=aire, tipo='condensador')

        # instancia ciclo cascada con estos HX
        ciclo_combinado = VCR_cascada(Q=Q,
                                     T_ev_low=T_e_low, T_cond_low=T_e_low + T_dif,
                                     T_ev_high=T_e_high, T_cond_high=T_c_high,
                                     ref1=fluid_low, ref2=fluid_high,
                                     eta_c_low=eta_s, eta_c_high=eta_s,
                                     Tsh=Tsh, Tsc=Tsc, T_amb=T_amb,
                                     T_chamber=T_chamber, m_g_e_low=m_g_e_low, m_aire=m_aire,
                                     ref_e_low=glicol, U_e_low=U_e_low,
                                     HX_e_low=HX_ev_low, HX_c_high=HX_cond_high)

        # En caso de que tu clase calcule valores en init y solve, ya estn listos
        COP = ciclo_combinado.COP
        costo = ciclo_combinado.costo
        Ex_d_total = ciclo_combinado.Ex_d_total

        # extrae otras variables para restricciones / diagnstico
        T_lm_ev = ciclo_combinado.T_lm_e_low
        T_lm_cond = ciclo_combinado.T_lm_c_high
        m_dot_low = ciclo_combinado.m_dot_low
        m_dot_high = ciclo_combinado.m_dot_high
        Pe_low = ciclo_combinado.Pe_low
        Pc_low = ciclo_combinado.Pc_low
        Pe_high = ciclo_combinado.Pe_high
        Pc_high = ciclo_combinado.Pc_high

        A_e_low = ciclo_combinado.HX_e_l.A
        A_c_high = ciclo_combinado.HX_c_h.A
        A_HX = ciclo_combinado.A_HX

        out ={ 
            "COP": COP,
            "Ex_dest": Ex_d_total,
            "costo": costo,
            "T_lm_ev": T_lm_ev,
            "T_lm_cond": T_lm_cond,
            "m_dot_low": m_dot_low,
            "m_dot_high": m_dot_high,
            "Pe_low": Pe_low,
            "Pc_low": Pc_low,
            "Pe_high": Pe_high,
            "Pc_high": Pc_high,
            "A_e_low": A_e_low,
            "A_c_high": A_c_high,
            "A_HX": A_HX,
            "ok": True,
        }

    except Exception as e:
        out["ok"] = False

    return out

class ciclo_multi(ElementwiseProblem):
    def __init__(self):
        super().__init__(
            n_var=3, #Pinch en el ev. y cond. + Tdif cascada
            n_obj=3,  #Funcin objetivo: minimizar Ex_dest y C_total y maximizar COP
            n_constr=8, #
            xl=np.array([0.1, 0.1, 25]),
            xu=np.array([10, 10, 35])
        )
    
    
    def _evaluate(self, x, out, *args, **kwargs):
        solucion = componentes(x[0], x[1], x[2])
        COP = solucion["COP"]
        costo = solucion["costo"]
        Ex_d = solucion["Ex_dest"]

        m_dot_l = solucion["m_dot_low"]
        m_dot_h = solucion["m_dot_high"]
        Pc_low = solucion["Pc_low"]
        Pe_low = solucion["Pe_low"]
        Pc_high = solucion["Pc_high"]
        Pe_high = solucion["Pe_high"]
        Ae_low = solucion["A_e_low"]
        Ac_high = solucion["A_c_high"]
        AHX = solucion["A_HX"]


        #Restricc g<=0
        g1 = -m_dot_l
        g2 = -m_dot_h

        # Relacin de compresin  8-10 para eficiencia y seguridad
        rc_low = Pc_low / Pe_low
        rc_high = Pc_high / Pe_high
        #g8 = rc_low - 50.0  # rc_low  10
        #g9 = rc_high - 50.0  # rc_high  10
        # reas mnimas y mximas para intercambiadores
        g10 = 0.1 - Ae_low    # A_evap  0.1 m
        g11 = 0.1 - Ac_high   # A_cond  0.1 m  
        g12 = 0.1 - AHX       # A_HX  0.1 m

        g13 = Ae_low - 200     # A_evap  200 m
        g14 = Ac_high - 200    # A_cond  200 m
        g15 = AHX - 200        # A_HX  200 m

        f1 = -COP
        f2 = costo
        f3 = Ex_d
        out["F"] = [f1, f2, f3]
        out["G"] = [g1, g2, g10, g11, g12, g13, g14, g15]
```

```python
problema = ciclo_multi()
algoritmo = NSGA2(pop_size=100)

solucion = minimize(problema,
               algoritmo,
               ('n_gen', 50),
               seed=1,
               verbose=True)
```

## Anlisis de Variables del Sistema vs Funciones Objetivo

A continuacin se presentan grficos de cada variable del sistema (df_vo) en funcin de las tres funciones objetivo: COP, Costo y Exerga Destruida (df_fo).

```python
F = solucion.F
X = solucion.X

df_fo = pd.DataFrame(F, columns=['-COP', 'Costo (USD/ao)', 'Ex_d (W)'])
df_fo["-COP"] = -df_fo['-COP']
df_fo.rename(columns={'-COP': 'COP'}, inplace=True)

df_fo["Costo (USD/ao)"] = df_fo["Costo (USD/ao)"]/1000  #kUSD/ao
df_fo.rename(columns={'Costo (USD/ao)': 'Costo (kUSD/ao)'}, inplace=True)

df_fo["Ex_d (W)"] = df_fo["Ex_d (W)"]/1000  #kW
df_fo.rename(columns={'Ex_d (W)': 'Ex_d (kW)'}, inplace=True)

df_vo = pd.DataFrame(X, columns=["pinch_e","pinch_cond", "T_casc" ])

variables = {"COP": [], "Ex_dest" : [], "costo": [] , "T_lm_ev": [], "T_lm_cond": [], "m_dot_low": [], "m_dot_high": [], "Pe_low": [], "Pc_low": [], "Pe_high": [], "Pc_high": [], "A_e_low": [], "A_c_high": [], "A_HX": [] }
for i in range(len(df_vo)):
    resultado = componentes(df_vo.loc[i, "pinch_e"], df_vo.loc[i, "pinch_cond"], df_vo.loc[i, "T_casc"])
    for key in variables.keys():
        variables[key].append(resultado[key])
for key in variables.keys():
    df_vo[key] = variables[key]
```

```python
COP_max, COP_min = max(df_fo["COP"]), min(df_fo["COP"])
i_max = df_fo["COP"].idxmax()
i_min = df_fo["COP"].idxmin()

p_2 = np.array((COP_min, df_fo.loc[i_min, "Costo (kUSD/ao)"], 0))
p_1 = np.array((COP_max, df_fo.loc[i_max, "Costo (kUSD/ao)"], 0))

def dis_p(p):
    num = np.cross((p - p_1), (p_2 - p_1))
    den = np.linalg.norm(p_2 - p_1)
    d = np.abs(num[2]) / den
    return d 

p_point = []
dist_p = []

for i in range(len(df_fo)):
    p = np.array((df_fo.loc[i, "COP"], df_fo.loc[i, "Costo (kUSD/ao)"], 0))
    d = dis_p(p)
    p_point.append(p)
    dist_p.append(d)


knee_idx = np.argmax(dist_p)
print(f"Knee point encontrado en ndice: {knee_idx}")
print(f"COP: {df_fo.loc[knee_idx, 'COP']:.3f}")
print(f"Costo: {df_fo.loc[knee_idx, 'Costo (kUSD/ao)']:.3f} kUSD/ao")
print(f"Ex_d: {df_fo.loc[knee_idx, 'Ex_d (kW)']:.3f} kW")

plt.figure(figsize=(10,6))
plt.scatter(df_fo["COP"], df_fo["Costo (kUSD/ao)"], c=df_fo["Ex_d (kW)"], cmap='viridis', alpha=0.6)
plt.colorbar(label='Ex_d (kW)')


plt.scatter(df_fo.loc[knee_idx, "COP"], df_fo.loc[knee_idx, "Costo (kUSD/ao)"], 
            color='red', s=200, marker='*', edgecolors='black', linewidths=2, 
            label=f'Knee Point (idx={knee_idx})', zorder=5)

plt.scatter(COP_max, df_fo.loc[i_max, "Costo (kUSD/ao)"], 
            color='blue', s=100, marker='o', edgecolors='black', linewidths=1.5, 
            label='COP Mximo', zorder=4)
plt.scatter(COP_min, df_fo.loc[i_min, "Costo (kUSD/ao)"], 
            color='green', s=100, marker='o', edgecolors='black', linewidths=1.5, 
            label='COP Mnimo', zorder=4)

plt.plot([COP_min, COP_max], [df_fo.loc[i_min, "Costo (kUSD/ao)"], df_fo.loc[i_max, "Costo (kUSD/ao)"]], 
         'k--', alpha=0.3, linewidth=1.5, label='Lnea de Referencia')

plt.xlabel("COP")
plt.ylabel("Costo (kUSD/ao)")
plt.title("Frontera de Pareto con Knee Point")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Despus de la optimizacin, extraer el frente Pareto
def save_pareto_front(result, filename="pareto_front.csv"):
    # Puntos del frente Pareto
    F = result.F
    X = result.X
    
    pareto_data = []
    for i in range(len(F)):
        # Reclculo para obtener mtricas completas
        sol = componentes(X[i,0], X[i,1], X[i,2])
        if sol["ok"]:
            pareto_data.append({
                'pinch_ev': X[i,0],
                'pinch_cond': X[i,1], 
                'T_dif': X[i,2],
                'COP': -F[i,0],  # Porque minimizabas -COP
                'costo': F[i,1],
                'Ex_dest': F[i,2],
                'm_dot_low': sol["m_dot_low"],
                'm_dot_high': sol["m_dot_high"],
                'A_e_low': sol["A_e_low"],
                'A_c_high': sol["A_c_high"],
                'A_HX': sol["A_HX"],
                'rank': i  # Ranking en el frente Pareto
            })
    
    df_pareto = pd.DataFrame(pareto_data)
    #df_pareto.to_csv(filename, index=False)
    return df_pareto

# Usar
df_pareto = save_pareto_front(solucion)

COP_max, COP_min = max(df_pareto["COP"]), min(df_pareto["COP"])
i_max = df_pareto["COP"].idxmax()
i_min = df_pareto["COP"].idxmin()

p_2 = np.array((COP_min, df_pareto.loc[i_min, "costo"], 0))
p_1 = np.array((COP_max, df_pareto.loc[i_max, "costo"], 0))

def dis_p(p):
    num = np.cross((p - p_1), (p_2 - p_1))
    den = np.linalg.norm(p_2 - p_1)
    d = np.abs(num[2]) / den
    return d 

p_point = []
dist_p = []

for i in range(len(df_pareto)):
    p = np.array((df_pareto.loc[i, "COP"], df_pareto.loc[i, "costo"], 0))
    d = dis_p(p)
    p_point.append(p)
    dist_p.append(d)


knee_idx = np.argmax(dist_p)
print(f"Knee point encontrado en ndice: {knee_idx}")
print(f"COP: {df_pareto.loc[knee_idx, 'COP']:.3f}")
print(f"Costo: {df_pareto.loc[knee_idx, 'costo']:.3f} kUSD/ao")
print(f"Ex_d: {df_pareto.loc[knee_idx, 'Ex_dest']:.3f} kW")

plt.figure(figsize=(10,6))
plt.scatter(df_pareto["COP"], df_pareto["costo"], c=df_pareto["Ex_dest"], cmap='viridis', alpha=0.6)
plt.colorbar(label='Ex_dest')


plt.scatter(df_pareto.loc[knee_idx, "COP"], df_pareto.loc[knee_idx, "costo"], 
            color='red', s=200, marker='*', edgecolors='black', linewidths=2, 
            label=f'Knee Point (idx={knee_idx})', zorder=5)

plt.scatter(COP_max, df_pareto.loc[i_max, "costo"], 
            color='blue', s=100, marker='o', edgecolors='black', linewidths=1.5, 
            label='COP Mximo', zorder=4)
plt.scatter(COP_min, df_pareto.loc[i_min, "costo"], 
            color='green', s=100, marker='o', edgecolors='black', linewidths=1.5, 
            label='COP Mnimo', zorder=4)

plt.plot([COP_min, COP_max], [df_pareto.loc[i_min, "costo"], df_pareto.loc[i_max, "costo"]], 
         'k--', alpha=0.3, linewidth=1.5, label='Lnea de Referencia')

plt.xlabel("COP")
plt.ylabel("costo")
plt.title("Frontera de Pareto con Knee Point [R41]")
plt.legend()
plt.grid(True)
plt.show()

#3 Variables decision
#3 Areaas
pinch_ev_knee = df_pareto.loc[knee_idx, "pinch_ev"]
pinch_cond_knee = df_pareto.loc[knee_idx, "pinch_cond"]
T_dif_knee = df_pareto.loc[knee_idx, "T_dif"]

area_ev_knee = df_pareto.loc[knee_idx, "A_e_low"]
area_cond_knee = df_pareto.loc[knee_idx, "A_c_high"]
area_inter = df_pareto.loc[knee_idx, "A_HX"]

print(f"Pinch Evaporador: {pinch_ev_knee} K\nPinch Condensador: {pinch_cond_knee} K\nDiferencia de Temperatura Cascada: {T_dif_knee} K\nrea Evaporador: {area_ev_knee} m\nrea Condensador: {area_cond_knee} m\nrea Intercambiador de Cascada: {area_inter} m")
```
